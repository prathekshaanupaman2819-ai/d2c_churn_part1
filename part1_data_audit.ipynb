```python
import pandas as pd

# Now that all files are in the folder, just load them by name!
customers = pd.read_csv('customers.csv')
orders = pd.read_csv('orders.csv')
tickets = pd.read_csv('support_tickets.csv')

# Merge them: Add orders and tickets to your customer list

master_df = customers.merge(orders, on='customer_id', how='left')
master_df = master_df.merge(tickets, on='customer_id', how='left')


print("Total records in your Master Table:", len(master_df))
print("Missing values in your Master Table:")
print(master_df.isnull().sum())
```

**​1. Executive Summary**

​The purpose of this audit was to assess the reliability and completeness of the customer data for our upcoming churn prediction project. The dataset has been validated for uniqueness and consistency, confirming it is suitable for further analysis.

**​2. Methodology**

​The data audit was conducted using Python and the pandas library within Google Colab. I merged the customer, order, and support ticket files on the customer_id key to create a comprehensive "Master Table" for analysis.

**​3. Key Findings**

​Data Integrity: I performed a duplication check on the customer_id column and confirmed 0 duplicate records.

​Data Volume: The Master Table contains 10,009 total orders, providing a strong foundation for our modeling efforts.

​Completeness: My analysis of missing values identified that fields such as loyalty_tier and skin_type contain significant missing data, which will need to be addressed before building the prediction model.

**​4. Recommendations**

​Missing Data Handling: I recommend treating missing values in the loyalty_tier and skin_type categories as a distinct "Unknown" category rather than dropping these rows, as this will preserve our sample size.

​Ongoing Validation: Future data imports should be checked against the customer_id master list to ensure there are no orphaned records in the order or support ticket logs.
